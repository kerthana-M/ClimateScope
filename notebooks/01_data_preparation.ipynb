```python
import pandas as pd
import numpy as np
```

```python

data_path = "../data/raw/GlobalWeatherRepository.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
print("Dataset shape:", df.shape)
print("\nColumns:")
print(df.columns)
df.info()
```

```python
missing_values = df.isnull().sum()
```

```python
# Filling numerical missing values with column mean
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Filling categorical missing values with 'Unknown'
categorical_cols = df.select_dtypes(include=["object"]).columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")

# Verifying missing values again
df.isnull().sum()
```

```python
# Converting timestamp column to datetime
df["last_updated"] = pd.to_datetime(df["last_updated"], errors="coerce")

# Setting datetime column as index
df.set_index("last_updated", inplace=True)

# Extracted year and month for aggregation
df["year"] = df.index.year
df["month"] = df.index.month
df.head()
```

```python
df.describe()
```

```python
# cleaned dataset is saved 
processed_path =  "../data/processed/weather_cleaned.csv"

df.to_csv(processed_path, index=False)

print("Cleaned dataset saved to:", processed_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["temperature_celsius"], bins=40, kde=True)
plt.title("Distribution of Temperature (°C)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```

```python
monthly_temp = (
    df.groupby(["year", "month"])["temperature_celsius"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(8,4))
sns.lineplot(
    data=monthly_temp,
    x="month",
    y="temperature_celsius",
    ci=None
)
plt.title("Average Monthly Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Average Temperature (°C)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["precip_mm"], bins=40)
plt.title("Precipitation Distribution (mm)")
plt.xlabel("Precipitation (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Identify Seasons

def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

df["season"] = df["month"].apply(get_season)

print("Season column added")
df[["year", "month", "season"]].head()
```

```python
# Country-wise Seasonal Analysis
# Select a country (change if needed)
country_name = "India"
country_df = df[df["country"] == country_name]

print(f"Filtered data for country: {country_name}")
```

```python
#  Seasonal Temperature Visualization

seasonal_temp = (
    country_df.groupby("season")["temperature_celsius"]
    .mean()
    .reindex(["Winter", "Spring", "Summer", "Autumn"])
    .reset_index()
)

plt.figure(figsize=(5, 3))
sns.barplot(
    data=seasonal_temp,
    x="season",
    y="temperature_celsius"
)
plt.title(f"Average Seasonal Temperature in {country_name}")
plt.xlabel("Season")
plt.ylabel("Average Temperature (°C)")
plt.show()
```

```python
#  Seasonal Precipitation Visualization

seasonal_rain = (
    country_df.groupby("season")["precip_mm"]
    .mean()
    .reindex(["Winter", "Spring", "Summer", "Autumn"])
    .reset_index()
)

plt.figure(figsize=(5, 3))
sns.barplot(
    data=seasonal_rain,
    x="season",
    y="precip_mm"
)
plt.title(f"Average Seasonal Precipitation in {country_name}")
plt.xlabel("Season")
plt.ylabel("Precipitation (mm)")
plt.show()

```

```python
# Save Final Dataset with Seasons

final_path = "../data/processed/weather_cleaned_with_seasons.csv"
df.to_csv(final_path, index=False)

print("Final dataset with seasons saved to:", final_path)
```
