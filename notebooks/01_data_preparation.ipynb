```python
import pandas as pd
import numpy as np
```

```python
# Path to the raw dataset
data_path = "C:\\Users\\keert\\OneDrive\\Desktop\\ClimateScope\\data\\raw\\GlobalWeatherRepository.csv"

# Read CSV file
df = pd.read_csv(data_path)

# Display first 5 rows
df.head()
```

```python
# Shape of the dataset
print("Dataset shape:", df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Data types and missing values
df.info()
```

```python
# Count missing values per column
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values[missing_values > 0]
```

```python
# Fill numerical missing values with column mean
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill categorical missing values with 'Unknown'
categorical_cols = df.select_dtypes(include=["object"]).columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")

# Verify missing values again
df.isnull().sum()
```

```python
# Convert date column to datetime
df["timezone"] = pd.to_datetime(df["timezone"], errors="coerce")

# Set date as index (optional but recommended)
df.set_index("timezone", inplace=True)

df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Save processed data
processed_path =  "C:\\Users\\keert\\OneDrive\\Desktop\\ClimateScope\\data\\raw\\GlobalWeatherRepository.csv"
df.to_csv(processed_path)

print("Cleaned dataset saved to:", processed_path)
```
